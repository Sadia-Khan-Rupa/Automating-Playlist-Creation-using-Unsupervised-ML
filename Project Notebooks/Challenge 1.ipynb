### 5.2.&nbsp; Use KMeans to group the 10 songs dataset

Play around with the clustering algorithm and get familiar with it. What happens when you change the number of clusters? What number of clusters produces the best output? What happens when you change the number of the random seed?

```python
#import the model
from sklearn.cluster import KMeans
```

```python
#number of clusters = 3, random state = None
myKmeans = KMeans(n_clusters=3, random_state=None)
myKmeans.fit(songs_df)
```

```python
cluster = myKmeans.labels_
cluster
```

```python
songs_df['cluster'] = cluster
songs_df[['cluster']].sort_values(by= 'cluster')
```

```python
#number of clusters = 4, random state = None
myKmeans_4_cluster = KMeans(n_clusters=4, random_state=None)
myKmeans_4_cluster.fit(songs_df)
cluster_4 = myKmeans_4_cluster.labels_
songs_df['cluster_4'] = cluster_4
songs_df[['cluster_4']].sort_values(by= 'cluster_4')
```

```python
#number of clusters = 4, random state = 0
myKmeans_4_cluster_rs0 = KMeans(n_clusters=4, random_state=0)
myKmeans_4_cluster_rs0.fit(songs_df)
cluster_4_0 = myKmeans_4_cluster_rs0.labels_
songs_df['cluster_4_0'] = cluster_4_0
songs_df[['cluster_4_0']].sort_values(by= 'cluster_4_0')
```

**when random state = none, Space Raiders - Charlotte de Witte Remix was in cluster 2, when random state = 0, Space Raiders - Charlotte de Witte Remix in cluster 1**

```python
#number of clusters = 4, random state = 42
myKmeans_4_cluster_rs42 = KMeans(n_clusters=4, random_state=42)
myKmeans_4_cluster_rs42.fit(songs_df)
cluster_4_42 = myKmeans_4_cluster_rs42.labels_
songs_df['cluster_4_42'] = cluster_4_42
songs_df[['cluster_4_42']].sort_values(by= 'cluster_4_42')
```
